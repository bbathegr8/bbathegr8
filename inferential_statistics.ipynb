###3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?
"- State the null and alternate hypotheses - Conduct the hypothesis test and compute the p-value - Write down conclusions from the test results Note: Consider the level of significance as 5%."

Null Hypothesis(H0) : The mean printing quality using unpolished stones is equal to or better than the mean printing quality using polished stones.

Alternative Hypothesis(H1) : The mean printing quality using unpolished stones is worse than the mean printing quality using polished stones.

````python
from scipy.stats import ttest_1samp
t_statistic, p_value = ttest_1samp(df1['Unpolished '], 134.110,alternative='less')
print("P-value", p_value)

# Check significance level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence to support Zingaro's belief that the printing quality using unpolished stones is worse than that using polished stones.")
else:
    print("Fail to reject the null hypothesis. There is insufficient evidence to conclude that the printing quality using unpolished stones is worse than that using polished stones.")
````

Null Hypothesis (H0)

1) The hardness of implants does not vary significantly depending on the dentist.
2) The hardness of implants does not vary significantly depending on the method.
3) There is no significant interaction effect between dentist and method on the hardness of implants.

Alternate Hypothesis(H1)

1) The hardness of implants varies significantly depending on the dentist.
2) The hardness of implants varies significantly depending on the method.
3) There is a significant interaction effect between dentist and method on the hardness of implants.

````python
formula = 'Response ~ C(Dentist)+C(Method)+C(Dentist):C(Method)'
model = ols(formula, Alloy1).fit()
aov_table = anova_lm(model)
(aov_table)
````
