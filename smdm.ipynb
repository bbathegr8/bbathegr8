```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
## Check the lower range and upper range for the variable with zscore
lr,ur=remove_outlier(df1['No_of_Dependents'])
print("lower range",lr, "and upper range", ur)
```

```python
df1['No_of_Dependents']=np.where(df1['No_of_Dependents']>ur,ur,df1['No_of_Dependents'])
df1['No_of_Dependents']=np.where(df1['No_of_Dependents']<lr,lr,df1['No_of_Dependents'])
```
